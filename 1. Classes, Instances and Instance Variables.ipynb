> 1. **Class** >      Blueprint for creating Instances.               
> 2. **Attributes** > Variable associated w/ a Class.                                      
> 3. **Methods** >    Functions associated w/ a Class.

```python
class Employee:
    pass

emp1 = Employee()
emp2 = Employee()
```

> 4. **Instance** > Objects *emp1* & *emp2* are two unique instances of class *Employee*.  
> 5. **Instance Variable** > Variables containing data unique to each instance.

```python
emp1.first = 'Joe'
emp1.last = 'Rogan'
emp1.email = 'joe_rogan@gmail.com'

emp2.first = 'Joe'
emp2.last = 'Biden'
emp2.email = 'joe_biden@gmail.com'

print(emp1.last)
print(emp2.email)
print(emp1.pay)
```

##### each of the instances above have attributes unque to them

```python
# when we create methods in class, they recieve the instance as the first argument automatically

class Employee:
    def __init__(self, first, last, pay):
        self.first = first
        self.last = last
        self.pay = pay
        self.email = first + '_' + last + '@gmail.com'
        
emp1 = Employee('Joe', 'Rogan', 50000)  # instance is passed automatically
emp2 = Employee('Joe', 'Biden', 40000)

print(emp1.last)
print(emp2.email)
```

```python
class Employee:
    
    def __init__(self, first, last, pay):
        self.first = first
        self.last = last
        self.pay = pay
        self.email = first + '_' + last + '@gmail.com'
        
    def full_name(self):
        return self.first + ' ' + self.last
    
emp1 = Employee('Joe', 'Rogan', 50000)  # instance is passed automatically
emp2 = Employee('Joe', 'Biden', 40000)

print(emp1.full_name())  # instance emp1 is automatically passed inside .full_name()
print(Employee.full_name(emp1))
```
